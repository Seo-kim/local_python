# 1.3 크롤링 csv 파일 통합 Preprocessing
> 목적 : 키워드별 // 언론사별 // 기간별 //로 수집된 csv파일들에 대해
=> 입력값에 대해 전처리를 수행하고   
=> 필요에 맞게 하나의 파일로 merge한다.
  
  
> 작업 :  
=> date, text 값들이 오류없이 잘 들어 있는지 체크   
  => 키워드별 // 언론사 + 기간 통합된 하나의 csv로 merge하기.  
  => ex) all_공천.csv : '기사URL', '기사제목', '댓글날짜', '댓글원문', '언론사' 가 보존되어 있어야 한다.

## 언론3사 > 20대 국회_총합.csv 제작

```python
total = pd.read_csv('../KBS시사/언론3사_20대/all_20.csv', engine='python', header=0, index_col = 0)
# total.reset_index(drop = True, inplace = True)
print(len(total))
total.drop_duplicates(inplace = True)
print(len(total))
```

### 조선일보 URL 붙이기

```python
total = pd.read_csv('../KBS시사/언론3사_20대/all_20대.csv', engine='python', header=0)
total.head()
```

```python
total_chosun = total.loc[total['from'] == 'Chosun',].copy()
```

```python
total_chosun['title'] = list(map(lambda x: x.replace('<U+22EF>','⋯').replace('<U+22C5>','⋅') , total_chosun['title'].tolist())) 
# <U+22C5> => ⋅
# <U+22EF> => ⋯
```

```python
file_list = glob('../KBS시사/DaumPoli/조선일보_전체/조선_20대_2019리플수집/Chosun_20대 국회_post_*.xlsx')
```

```python
all_df = pd.DataFrame()

for file_name in file_list:
    appdx_df = pd.read_excel(file_name)
    all_df = pd.concat([all_df,appdx_df],ignore_index = True)
    
```

```python
total_chosun = total_chosun.merge(all_df, on ='title', how = 'left')
total_chosun.head(10)
```

```python
len(total_chosun.link.unique())  # total_df에는 358개 기사에 url이 붙었고
```

```python
len(all_df.link.unique()) # 전체 조선일보 갯수는 400개였다. url이 안붙은게 42개가 있는데,
```

```python
etc_titles = pd.concat([total_chosun[['title','link']], all_df[['title','link']] ]).drop_duplicates(keep = False).dropna().reset_index(drop = True).title
print(etc_titles)
```

```python
total_chosun.loc[total_chosun['link'].isnull(),].title.unique()
# <U+22EF> => ⋯
# <U+22C5> => ⋅
# total_df 에 있는 조선일보 리플에는 url이 최소한 다 붙긴 붙었다. 
# >>> total_df생성과정중에 42개 기사에 대한 리플이 누락된 것인지 파악 필요. 
```
